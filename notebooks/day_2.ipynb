```python
from account.savings_account import SavingsAccount
from account.premium_account import PremiumAccount
from account.investment_account import InvestmentAccount, Asset
from account.enums import AssetType

savings1 = SavingsAccount(
    owner_name="Alice",
    id="SAV-1",
    balance=1000,
    min_balance=200,
    monthly_interest=2
)

savings2 = SavingsAccount(
    owner_name="Bob",
    id="SAV-2",
    balance=500,
    min_balance=100,
    monthly_interest=1.5
)

savings1.deposit(200)
savings1.apply_monthly_interest()
try:
    savings2.withdraw(450)
except Exception as e:
    print("Savings withdraw error:", e)
print(savings1)
print(savings2)
# --- Premium Accounts ---
premium1 = PremiumAccount(
    owner_name="Charlie",
    id="PREM-1",
    balance=1000,
    fee=5,
    overdraft_limit=500
)

premium2 = PremiumAccount(
    owner_name="Diana",
    id="PREM-2",
    balance=300,
    fee=10,
    overdraft_limit=200
)

premium1.withdraw(1200)  # работает благодаря overdraft
premium2.deposit(200)

print(premium1)
print(premium2)

assets1 = [
    Asset(AssetType.STOCK, "Apple", 1000, 8),
    Asset(AssetType.ETF, "S&P 500 ETF", 2000, 7),
]

assets2 = [
    Asset(AssetType.BOND, "US Treasury", 1500, 3),
    Asset(AssetType.STOCK, "Tesla", 1200, 10),
]

invest1 = InvestmentAccount(
    assets=assets1,
    owner_name="Eve",
    id="INV-1",
    balance=5000
)

invest2 = InvestmentAccount(
    assets=assets2,
    owner_name="Frank",
    id="INV-2",
    balance=3000
)

print(invest1)
print("Projected yearly growth:", invest1.project_yearly_growth())

print(invest2)
print("Projected yearly growth:", invest2.project_yearly_growth())
```
