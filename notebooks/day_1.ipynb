```python
import sys
sys.path.append('../src')

from account.bank_account import BankAccount
from account.enums import AccountStatus
from account.base import InvalidOperationError, InsufficientFundsError, AccountFrozenError

account = BankAccount('So1iD')
frozen_account = BankAccount(owner_name='So1iD', status=AccountStatus.FROZEN)

print(account)
print('deposit 200')
account.deposit(200)
print(account)
print('withdraw 100')
account.withdraw(100)
print(account.get_account_info())

try:
    account.deposit(-200)
except InvalidOperationError:
    print('InvalidOperationError')

try:
    account.withdraw(-200)
except InvalidOperationError:
    print('InvalidOperationError')

try:
    account.withdraw(2000)
except InsufficientFundsError:
    print('InsufficientFundsError')

print(frozen_account)
try:
    frozen_account.deposit(2000)
except AccountFrozenError:
    print('AccountFrozenError')

print(account.get_account_info())
```
